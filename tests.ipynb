```python
import datetime as dt

import json
```

```python
ficheroWrite = open("Lista_Intervalos.json", "w+")
ficheroAppend = open("Lista_Intervalos.json","a")
```

```python
 def crear_intervalo():
        ##n = len(intervalos_lista)
        interval = {
            ##"id": n,
            "inicio": dt.datetime.today().isoformat(timespec="seconds"),
            "final": None, # final del intervalo,
            "duracion": None,# duracion del intervalo,
        }
        print("Se ha creado un intervalo: "+ str(interval))
        return interval
```

```python
crear_intervalo()
```

```python
lista_intervalos = []
```

```python
with open("Lista_Intervalos.json", "w+") as f:
    lista_intervalos.append(crear_intervalo())

    json.dump(lista_intervalos, f, indent=4, ensure_ascii=False)
    ficheroWrite.close()
```
