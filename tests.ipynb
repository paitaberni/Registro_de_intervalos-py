```python
 def crear_intervalo():
        ##n = len(intervalos_lista)
        interval = {
            ##"id": n,
            "inicio": dt.datetime.today().isoformat(timespec="seconds"),
            "final": None, # final del intervalo,
            "duracion": None,# duracion del intervalo,
        }
        print("Se ha creado un intervalo: "+ str(interval))
        return interval
```

```python

    def Leer_Lista():
        with open("Lista_Intervalos.json", "r") as f:

            lista_intervalos = json.load(f)

        return lista_intervalos
```

```python
def cerrar_intervalo():
    lista_intervalos = Leer_Lista()
    print("Elige un intervalo para cerrar:")
    ##for i in range(len(lista_intervalos)):
      ##  print(str(i)+".  "+str(lista_intervalos[i]))

    indice = 0  ##int(input("Ingrese el nombre del intervalo: "))
    intervalo_seleccionado = lista_intervalos[indice]
    intervalo_seleccionado["final"] = dt.datetime.today()
    inicio = dt.datetime.fromisoformat(intervalo_seleccionado["inicio"])

    intervalo_seleccionado["duracion"] = intervalo_seleccionado["final"] - inicio
    intervalo_seleccionado["duracion"] = int(intervalo_seleccionado["duracion"].total_seconds())
    print("se ha modificado un intervalo: "+ str(intervalo_seleccionado))
    intervalo_seleccionado["final"] = intervalo_seleccionado["final"].isoformat(timespec="seconds")
    return intervalo_seleccionado
```

```python
lista_intervalos =  Leer_Lista()
print(lista_intervalos)
lista_intervalos.append(cerrar_intervalo())

with open("Lista_Intervalos.json", "w") as k:
        json.dump(lista_intervalos, k, indent=4, ensure_ascii=False)
        ficheroWrite.close()
```
